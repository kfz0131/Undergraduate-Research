```python
import nltk
from nltk.corpus import reuters
import pandas as pd
import pprint
pp = pprint.PrettyPrinter(indent=2, depth=2, width=80, compact=True)
```

```python
# Categories in reuters article
print('Total number of categories in reuters articles = {0:d}'.format(len(reuters.categories())))
print(75*'-')
pp.pprint(reuters.categories())
```

```python
# Dictionary of category with how many times it appears 
category_dic = {cat : len(reuters.fileids(cat)) for cat in reuters.categories()}
print('Number of times each category appears :')
print(75*'-')
pp.pprint(category_dic)
```

```python
# What are those articles specifically
category_exactdic = {cat : reuters.fileids(cat) for cat in reuters.categories()}
print('Articles that are tagged with the same category :')
print(75*'-')
pp.pprint(category_exactdic)
```

```python
# Get a general idea of how many categories are tagged to each article
article_dic = {art : len(reuters.categories(art)) for art in reuters.fileids()}
print('Number of category tags for each article :')
print(75*'-')
pp.pprint(article_dic)
```

```python
# What are those categories specifically
article_exactdic = {art : reuters.categories(art) for art in reuters.fileids()}
print('Category tags for each article :')
print(75*'-')
pp.pprint(article_exactdic)
```

```python
# Set up the matrix with ID as index and Categories as columns
category_matrix = pd.DataFrame(index = reuters.fileids(), columns = reuters.categories())
# Loop through all columns (categories)
for i in category_matrix.axes[1]:
    # Loop through all rows (articles) to check if this article is tagged with this category
    for j in category_matrix.axes[0]:
        if (i in reuters.categories(j)):
            category_matrix[i][j] = 1
```
